```python
import os
from google.oauth2 import service_account
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload

# --- Your Configuration ---
# Path to your service account key file
SERVICE_ACCOUNT_FILE = 'manifest.json'
# The file you want to upload
FILE_PATH = "output.wav"
# The name you want the file to have in Google Drive
DRIVE_FILE_NAME = "wavfiles/output.wav"
# The ID of the folder you shared with the service account
FOLDER_ID = "1TIXeTnGnYrVru8MfpzUtepBF6FLA1_KJ" # <--- IMPORTANT: PASTE YOUR FOLDER ID HERE

# --- Don't edit below this line ---

# Scopes define the level of access.
SCOPES = ['https://www.googleapis.com/auth/drive']

def authenticate():
    """Authenticates the service account and returns the service object."""
    creds = service_account.Credentials.from_service_account_file(
                SERVICE_ACCOUNT_FILE, scopes=SCOPES)
    return creds

# --- THIS IS THE CORRECTED FUNCTION ---
def upload_file(drive_service, file_path, drive_file_name, folder_id):
    """Uploads a file to the specified folder in Google Drive using a resumable upload."""
    try:
        print(f"Starting upload of {file_path}...")
        import mimetypes
        mimetype = mimetypes.guess_type(file_path)[0] or 'application/octet-stream'
            
        file_metadata = {
            'name': drive_file_name,
            'parents': [folder_id] # Specify the parent folder
        }
        
        # This is the corrected line.
        # It enables resumable uploads, sending the file in chunks.
        # 10 * 1024 * 1024 = 10MB chunk size. You can adjust this.
        media = MediaFileUpload(
            file_path, 
            mimetype=mimetype, 
            chunksize=10 * 1024 * 1024, 
            resumable=True
        )
        
        # The request object now includes the resumable upload logic
        request = drive_service.files().create(
            body=file_metadata,
            media_body=media,
            fields='id'
        )

        # Execute the upload and get progress
        response = None
        while response is None:
            status, response = request.next_chunk()
            if status:
                print(f"Uploaded {int(status.progress() * 100)}%")

        print(f"File uploaded successfully! File ID: {response.get('id')}")
        return response.get('id')
        
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


# if __name__ == '__main__':
#     if FOLDER_ID == "YOUR_FOLDER_ID_HERE":
#         print("Please edit the script and replace 'YOUR_FOLDER_ID_HERE' with your actual Google Drive folder ID.")
#     else:
#         creds = authenticate()
#         drive_service = build('drive', 'v3', credentials=creds)
#         upload_file(drive_service, FILE_PATH, DRIVE_FILE_NAME, FOLDER_ID)


from google.oauth2 import service_account
from googleapiclient.discovery import build

# --- Configuration ---
SERVICE_ACCOUNT_FILE = 'manifest.json'
SCOPES = ['https://www.googleapis.com/auth/drive']

# --- Authenticate ---
creds = service_account.Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE, scopes=SCOPES)
drive_service = build('drive', 'v3', credentials=creds)

# --- List all files ---
results = drive_service.files().list(
    pageSize=1000,  # Adjust as needed
    fields="files(id, name)").execute()
items = results.get('files', [])

if not items:
    print("No files found.")
else:
    print(f"Found {len(items)} files. Deleting...")

    for item in items:
        try:
            drive_service.files().delete(fileId=item['id']).execute()
            print(f"Deleted: {item['name']} ({item['id']})")
        except Exception as e:
            print(f"Failed to delete {item['name']} ({item['id']}): {e}")

print("Done.")
```

```python
from googleapiclient.http import MediaFileUpload
import os

def upload_folder(service, folder_path, parent_folder_id):
    """Uploads a folder and its contents to Google Drive, preserving structure."""
    folder_name = os.path.basename(folder_path.rstrip('/'))
    # Create the folder in Drive
    file_metadata = {
        'name': folder_name,
        'mimeType': 'application/vnd.google-apps.folder',
        'parents': [parent_folder_id]
    }
    folder = service.files().create(body=file_metadata, fields='id').execute()
    drive_folder_id = folder.get('id')

    for root, dirs, files in os.walk(folder_path):
        rel_path = os.path.relpath(root, folder_path)
        parent_id = drive_folder_id
        # Create subfolders if needed
        if rel_path != '.':
            # Find or create subfolder in Drive
            path_parts = rel_path.split(os.sep)
            parent_id = drive_folder_id
            for part in path_parts:
                query = f"'{parent_id}' in parents and name='{part}' and mimeType='application/vnd.google-apps.folder'"
                results = service.files().list(q=query, fields="files(id, name)").execute()
                items = results.get('files', [])
                if items:
                    parent_id = items[0]['id']
                else:
                    subfolder_metadata = {
                        'name': part,
                        'mimeType': 'application/vnd.google-apps.folder',
                        'parents': [parent_id]
                    }
                    subfolder = service.files().create(body=subfolder_metadata, fields='id').execute()
                    parent_id = subfolder.get('id')
        # Upload files in this directory
        for fname in files:
            file_path = os.path.join(root, fname)
            file_metadata = {
                'name': fname,
                'parents': [parent_id]
            }
            media = MediaFileUpload(file_path, resumable=True)
            service.files().create(body=file_metadata, media_body=media, fields='id').execute()

# Example usage:
upload_folder(drive_service, '/ocean/projects/cis250085p/shared/A_track/processed', FOLDER_ID)
```
